```python
import os
from sys import argv, stdout, stderr
from ROOT import TFile, gROOT, TCanvas, TLegend
import array
import sys                                                                                                                                                                                                                                                  
gROOT.SetStyle("Plain")                                                                                                                                                                                                                                           
gROOT.SetBatch(True)
```

```python
def positivize(histogram):
    output = histogram.Clone()
    for i in range(output.GetSize()):
        if output.GetArray()[i] < 0:
            output.AddAt(0, i)
    return output
```

```python
f = TFile('shapesMTBDT.root')
h0 = f.Get('0jet/Fakes')
print h0.Integral()
h1 = f.Get('0jet/Fakes_CMS_MuonFakeRate_p0_13TeVUp')
print h1.Integral()
h2 = f.Get('0jet/Fakes_CMS_MuonFakeRate_p0_13TeVDown')
print h2.Integral()
h1 = f.Get('0jet/Fakes_CMS_MuonFakeRate_p1_13TeVUp')
print h1.Integral()
h2 = f.Get('0jet/Fakes_CMS_MuonFakeRate_p1_13TeVDown')
print h2.Integral()
```

```python
canvas = TCanvas("", "", 800, 800)
h0 = positivize(h0)
binning = array.array('d', [-1.0, -0.5, 0.0, 0.5, 1.0])
h0 = h0.Rebin(len(binning)-1, 'LFVGG', binning)
print h0.Integral()
h0.Draw()
canvas.Draw()
```
