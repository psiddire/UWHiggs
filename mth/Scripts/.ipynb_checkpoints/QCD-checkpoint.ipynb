```python
import os
from sys import argv, stdout, stderr
from ROOT import TFile, gROOT, TCanvas, TLegend
import sys                                                                                                                                                                                                                                                  
gROOT.SetStyle("Plain")                                                                                                                                                                                                                                           
gROOT.SetBatch(True)
```

```python
canvas = TCanvas("canvas","canvas",800,800)
legend = TLegend(0.52,0.7,0.8,0.8)
f1 = TFile('../results/DataApril3/AnalyzeMuTauZTTHWW/QCD_Pt-20toInf_MuEnrichedPt15_TuneCP5_13TeV_pythia8_v14-v1.root')                                                                                                                                                                                                                              
h1 = f1.Get('TauLooseOS/m_t_Col')
f2 = TFile('TQCD.root')                                                                                                                                                                                                                              
h2 = f2.Get('QCD0JetCut')
h1.Scale(1/h1.Integral(), "width")
h2.Scale(1/h2.Integral(), "width")
h2.SetLineColor(2)
legend.AddEntry(h1, "Loose Selection")
legend.AddEntry(h2, "Tight Selection")
h1.Draw()
h2.Draw("SAME")
legend.Draw()
canvas.Draw()
```
