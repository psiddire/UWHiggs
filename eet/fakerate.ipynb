```python
import os
from sys import argv, stdout, stderr
from ROOT import TCanvas, TLegend, gROOT, TFile, gPad, TPad, TF1, gStyle, TGraphErrors
import sys
from array import array                                                                                                                                                            
import numpy as np
gROOT.SetStyle("Plain")
gROOT.SetBatch(True)
```

```python
gStyle.SetOptFit(1)
canvas = TCanvas('canvas','canvas',800,800)
grid = TPad("grid", "", 0, 0, 1, 1)
grid.Draw()
grid.cd()
fi = TFile('FakeRate2017/frEBDM0.root')
h = fi.Get('fakerate')
h.Draw()
gPad.Update()
graph = h.GetPaintedGraph()
h.SetTitle("|#eta|^{#tau} < 1.5 & Decay Mode 0; p_{T}^{#tau} (GeV); Tight-To-Loose")
graph.GetYaxis().SetTitleOffset(1.2)
graph.SetMinimum(0)
graph.SetMaximum(1.0)
gPad.Update()
f = TF1("Linear law","[0]+x*[1]", 0, 200)
f.SetLineColor(4)
f.SetLineStyle(0)
fitResult = h.Fit(f, "S")
steps = 500
x = np.linspace(10.0, 200.0, num=steps)                                                                                                                                    
y = np.array([fitResult.GetParams()[0] + fitResult.GetParams()[1]*x[i] for i in range(steps)])                                                                             
ey = np.array([fitResult.GetErrors()[0] + fitResult.GetErrors()[1]*x[i] for i in range(steps)])                                                                            
zeros = np.zeros(graph.GetN())                                                                                                                                             
interval = TGraphErrors(steps, x, y, zeros, ey)                                                                                                                            
interval.SetFillColor(2)                                                                                                                                                   
interval.SetFillStyle(3002)                                                                                                                                                
interval.Draw("e3 same")
canvas.Draw()
```
